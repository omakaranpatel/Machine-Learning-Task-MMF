```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
le = LabelEncoder()
```

```python
le = LabelEncoder()
columns = [ 'workclass', 'education', 'marital-status',
    'occupation', 'relationship', 'race',
    'sex', 'native-country', 'class']
for columns in columns:
    df_train[columns] = le.fit_transform(df_train[columns])
df_train
```

```python
le = LabelEncoder()
columns = [ 'workclass', 'education', 'marital-status',
    'occupation', 'relationship', 'race',
    'sex', 'native-country', 'class']
for columns in columns:
    df_test[columns] = le.fit_transform(df_test[columns])
df_test
```

```python
X_train = df_train.drop('class', axis=1)
y_train = df_train['class']
X_test = df_test.drop('class', axis=1)
y_test = df_test['class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_classifier = RandomForestClassifier(random_state=42)
rf_classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score,classification_report, roc_auc_score, roc_curve,confusion_matrix
```

```python
y_pred = rf_classifier.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
y_proba = rf_classifier.predict_proba(X_test)[:, 1]  # Probabilities for ROC
auc = roc_auc_score(y_test, y_proba)
print("AUROC (AUC):", auc)
```

```python
fpr, tpr, _ = roc_curve(y_test, y_proba)
plt.plot(fpr, tpr, label=f"AUC = {auc:.2f}")
plt.plot([0, 1], [0, 1], 'k--')  # random guess line
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.grid()
plt.show()
```

```python
df_export = df_test.copy()
df_export['predicted_class'] = y_pred
df_export['predicted_probability'] = y_proba
```

```python
import plotly.express as px

fig = px.treemap(
    df_export,  # the DataFrame
    path=['education', 'sex'],  # hierarchy: education > sex
    values=None,  # auto-count each group
    color='predicted_class',  # use prediction class to color
    color_continuous_scale='Viridis',
    title='Income Prediction Treemap by Education and Gender'
)
fig.show()
```
